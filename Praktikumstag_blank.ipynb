## Was ist der Origin of Replication (oriC)?
Unter einem oriC versteht man den Ort im Genom (Erbgut) eines Organismus, an dem die Replikation, also die Vervielfältiugung von DNA, ihren Anfang nimmt.     
Wie so oft in der Biologie gibt es nicht den "einen" oriC den man schlicht an ein und derselben Sequenz erkennen könnte, tatsächlich unterscheiden sich die oriCs unterschiedlicher Organsimen oft sehr stark in ihrer Beschaffenheit (und Anzahl!).

Im heutigen Praktikum konzentrieren wir unsere Studien hauptsächlich auf einen der bestbeschriebensten Replikationsursprünge: Den oriC von E. coli. 

Aber zunächst möchten wir unsere Suche natürlich etwas motivieren. Warum ist es interessant, den oriC eines Organismus zu kennen bzw. zu studieren?

## Der oriC von E. coli
E. coli ist "ein gramnegatives, säurebildendes und peritrich begeißeltes Bakterium, das normalerweise im menschlichen und tierischen Darm vorkommt." - und aufgrund seiner relativ einfachen Beschaffenheit und Möglichkeit zur Kultivierung ein äußerst beliebter Modellorganismus.
E. coli verfügt über ein recht kleines zirkuläres Genom mit nur einem einzigen oriC. Allerdings verbirgt sich hinter "recht klein" eine für uns Menschen im tagtäglichen Leben unvorstellbar große Zahl. Denn tatsächlich verfügt E. coli über ein Genom bestehend aus 5 Millionen Basenpaaren. In diesem nun naiv nach dem oriC zu suchen wäre equivalent dazu alle Harry Potter Bücher zu mischen und im Anschluss nach dem ersten Satz zu durchforsten - in einer Sprache, die man selbst nicht spricht.

##  Modellbildung: Ein naives Modell der DNA Replikation

Die folgende Abbildung zeigt ein naives Modell der DNA Replikation, wie man es schon in den 50er-Jahren nach der Entdeckung der DNA durch Watson und Crick hätte zeichnen können:



Wie hier abgebildet komplementieren die Nukleinbasen Adenin (A), Thymin (T) und Guanin (G) und Cytosin (C) einzelner Nukleotide einander, wenn die DNA in der von Watson und Crick postulierten Doppel-Helix Struktur vorliegt. 

Eines der einfachsten Modelle zur DNA Vervielfältigung könnte also vorsehen, dass die Doppel-Helix sich einfach entwindet, die Basenpaarung aufgelöst wird und sich freie Nukleotide einfach an die nun als einzelne Stränge vorliegende "Parent" DNA lagern würden, mit dem Ergebnis, dass wir von jedem Strang eine Kopie erstellt hätten - tatsächlich müssen wir heute nur ein Schulbuch aufschlagen, um festzustellen: Scheinbar ist die Replikation ein wenig komplizierter.

##  Ein etwas detaillierteres Modell der DNA Replikation
Die folgende Zeichnung liefert eine detailliertere  Darstellung des Erbguts von E. coli:



Tatsächlich geht die bakterielle Replikation in E. coli von einem einzigen oriC aus. Damit die Replikation stattfinden kann, müssen die Stränge der DNA zunächst getrennt und entspiralisiert werden.
Dieser Prozess wird von mehreren Proteinen bewerkstelligt, wie beispielsweise DnaA und die Helikase, die dazu beitragen, das die DNA sich entspiralisiert und die Replikationsgabel enstehen kann. 

Ausgehend vom oriC wandert die Replikationsmaschinerie das Genom also in zwei unterschiedliche Richtungen gleichzeitig ab, bis sie sich auf der anderen Seite am terC (replication terminus) treffen. So entsteht eine vollständige Kopie des Genoms. Anhand der "Wanderrichtungen" entlang der DNA Einzelstränge (half strands) unterscheiden wir die Stränge in Forward und Reverse half strands. Dabei beziehen sich die Begriffe "Forward" und "Reverse" auf die Leserichtung während der Translation.

Darüber hinaus wird auf dem Bild bereits ersichtlich, das die neuen DNA-Stränge in 5' -> 3' Richtung synthetisiert werden, dies ist dem Umstand geschuldet, dass die DNA Polymerase, also das Protein, das die neuen Stränge produziert, nur in diese Richtung synthetisieren kann.

Daraus ergibt sich, wie auf folgendem Bild gezeigt, eine Asymmetrie während der Replikation. Während anhand der Reverse half-strands kontinuierlich repliziert werden kann, ergibt sich im Fall der Forward half-strands eine diskontinuierliche Synthese. Tatsächlich öffnet sich die Replikationsgabel stets nur ein Stück weit und es werden anhand des Forward half-strands nur kurze Stücke neuer DNA in 5' -> 3' Richtung hergestellt, sog. Okazaki Fragmente.



Im Anschluss an die Replikation aller benötigten Fragmente werden diese dann schließlich ligiert und die Replikation endet mit einer vollständigen Kopie des zirkulären bakteriellen Genoms.

Nun ist es etwas unhandlich, auf diese Weise mit einer Genomsequenz zu hantieren. 
Um unser Vorhaben etwas angenehmer zu gestalten, verwenden wir nun eine Code-Bibliothek mit dem Namen Biofsharp. Diese besitzt bereits einige Grundfunkionalitäten im Umgang mit Biologischen Sequenzen.

```polyglot-notebook
#r "nuget: BioFSharp.Interactive, 2.0.0-preview.3"
open BioFSharp

let tpForward = BioArray.ofNucleotideString tpForwardRaw
let tpReverse = BioArray.ofNucleotideString tpReverseRaw
```

```polyglot-notebook
Seq.take 500 tpForward
```

Zunächst zählen wir die Anzahl einzelner Nukleotide im Forward half strand

```polyglot-notebook
let anzahlForward_A =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.A)
    |> Seq.length

let anzahlForward_T =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.T)
    |> Seq.length

let anzahlForward_G =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.T)
    |> Seq.length

let anzahlForward_C =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.T)
    |> Seq.length
```

```polyglot-notebook
let anzahlReverse_A =
    tpReverse 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.A)
    |> Seq.length

let anzahlReverse_T =
    tpReverse 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.T)
    |> Seq.length

let anzahlReverse_G = 0

let anzahlReverse_C = 0
```

## Erklärung unserer Beobachtung durch die Integration von weiterem Fachwissen

Unser Modell berücksichtigt bereits die Asymmetrie der Replikation bedingt durch die Syntheserichtung der DNA Polymerase. Dies hat unmittelbar zu Folge, dass sich der Forward half strand von dem diskontinuierlich abgelesen wird (Stichwort Okazaki Fragmente) einen wesentlich längeren Teil seines Lebens in einer einzelsträngigen Konfiguration aufhält.

Tatsächlich wurde zuvor beschrieben, dass einzelsträngige DNA wesentlich höhere Mutationsraten aufweist. Dies kann dadurch begründet werden, das ein bestimmter chemischer Prozess, die Deamination von Cytosin zu Thymin, 100-fach häufiger Auftritt, wenn die DNA als Einzelstrang vorliegt. Was wiederum bedeutet, dass zunächst ein Cytosin zu einem Thymin mutiert, welches nun mit einem A gepaart wird. In der Folge verringert sich die Anzahl der Cytosine auf dem Forward Strang und die Anzahl der Guanine auf dem Reverse Strang. Etwas verwirrt? Wir auch! ## Von einer vagen Idee zu einem klar formulierten Berechnungsproblem

Wie können wir also unsere bisherigen Modellierungsanstrengungen verwenden, um eine Strategie festzulegen, die uns dabei hilft den oriC in E. coli zu identifizieren?

Was wir bisher annehmen bzw. wissen:
1. E. coli verfügt über einen oriC.   
2. Die Replikation verläuft asymmetrisch, ausgehend vom oriC in beide Richtungen.  
3. Die Replikation endet auf der gegenüberliegenden Seite des Genoms, wenn beide Replikationsschleifen am terC aufeinandertreffen.  
4. Daraus folgt, das jeder neu synthetisierte half-strand zur Hälfte aus einem Forward und einem Reverse Strang besteht.  
5. Wir konnten empirisch zeigen, dass Forward und Reverse Stränge sich in der Guanine und Cytosine Zusammensetzung unterscheiden.  

Verknüpfen wir alle diese Beobachtungen, können wir zu dem Schluss kommen, dass es möglich ist den oriC zu identifizieren, wenn wir das Genom ablaufen und dabei die Guanin und Cytosin Mengen im Blick behalten.



Bewegen wir uns also Schritt für Schritt entlang des Genoms von E. coli, sollte der der oriC also dort zu finden sein, wo die Differenz(#Guanine - #Cytosin) ihren kleinsten Wert aufweist!

Natürlich ist das für uns als Menschen ein schwieriges Unterfangen, schließlich besteht das Genom von E. coli aus 4,6 Millionen Basenpaaren. Daher sieht der nächste Arbeitsschritt im Alltag eines Bioinformatikers vor, dass wir unser Problem so formulieren, dass wir es mit Hilfe eines Computer lösen können!

Das könnte ungefähr so aussehen:

> Das "Berechne-alle-Werte-für-die-Differenz(#Guanine-#Cytosin)-Problem:  
`Input: Eine DNA Sequenz genannt "Genome"`  
`Output: Alle beobachteten Differenzen i = 0 bis zu i = |Länge des Genoms - 1| (#Guanine-#Cytosin).`
