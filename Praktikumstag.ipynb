## Was ist der Origin of Replication (kurz: OriC)?
Unter einem OriC versteht man den Ort im Genom (Erbgut) eines Organismus an dem die Replikation, also die Vervielfältiugung von DNA, ihren Anfang nimmt. Wie so oft in der Biologie gibt es nicht den "einen" OriC den man schlicht an ein und der selben Sequenz erkennen könnte, tatsächlich unterscheiden sich die OriCs unterschiedlicher Organsimen oft sehr stark in ihrer Beschaffenheit (und Anzahl!).

Im heutigen Praktikum konzentrieren wir unsere Studien hauptsächlich auf einen der bestbeschriebensten Replikationsursprünge: Den OriC von E. coli. 

Aber zunächst möchten wir unsere Suche natürlich etwas motivieren. Warum ist es interessant den OriC eines Organismus zu kennen bzw. zu studieren?

## Der OriC von E. coli
E. coli ist "ein gramnegatives, säurebildendes und peritrich begeißeltes Bakterium, das normalerweise im menschlichen und tierischen Darm vorkommt." - und aufgrund seiner relativ einfachen Beschaffenheit und Möglichkeit zur Kultivierung ein äußerst beliebter Modellorganismus.
E. coli verfügt über ein recht kleines zirkuläres Genom mit nur einem einzigen OriC. Allerdings verbirgt sich hinter "recht klein" eine für uns Menschen im tagtäglichen Leben unvostellbar große Zahl. Denn tatsächlich verfügt E. coli über ein Genom bestehend aus 5 Millionen Basenpaaren. In diesem nun naiv nach dem OriC zu suchen wäre equivalent dazu alle Harry Potter Bücher zu mischen und im Anschluss nach dem ersten Satz zu durchforsten - in einer Sprache die man selbst nicht spricht.

##  Ein etwas detaillierteres Modell der DNA Replikation
Die folgende Zeichnung liefert eine detailiertere Darstellung des Erbguts von E coli:




Tatsächlich geht die bakterielle Replikation in E. coli von einem einzigen oriC aus. Wie wir hier erkennen können verfügt ein DNA Strang über eine Leserichtung von 5' nach 3' (gemäß der Leserichtung der Ribosomen während der Translation). Daraus ergibt sich, dass man die einzelnen DNA Stränge in sog. Forward- (in Translations-Leserichtung) und Reverse half-strands unterscheidet. 

Damit die Replikation stattfinden kann müssen die Stränge der DNA zunächst getrennt werden:



Diese Trennung wird von mehreren Proteinen bewerkstelligt, wie beispielsweise DnaA, das dazu eiträgt das die DNA sich entspriralisiert und die Replikationsgabel enstehen kann. Darüberhinaus wird auf dem Bild bereits ersichtlich, das die neuen DNA Stränge in 3' -> 5' Richtung synthetisiert werden, dies ist dem Umstand geschuldet, dass das die DNA Polymerase, also das Protein das die neuen Stränge produziert, nur in diese Richtung synthetisieren kann.

Daraus Ergibt sich, wie auf folgendem Bild gezeigt, eine Asymentrie während der Replikation. Währen die neuen reverse half-strands kontinuierlich erzeugt werden können, ergibt sich bei den neuen Forward half-strands eine diskontinuierliche Synthese. Tatsächlich öffnet sich die Replikationsgabel stehts ein Stück und es werden kurze Stücke neuer DNA in 3' -> 5' Richtung hergestellt, sog. Okazaki Fragmente.
 


Im Anschluss an die Replikation aller benötigten Fragmente werden diese dann schließlich ligiert und die Replikation Endet mit einer vollständigen Kopie des zirlkulären bakteriellen Genoms.

Zum Glück ist das gesamte Erbgut vieler Organismen wie E. coli bereits bekannt und öffentlich einsehbar. 
Darunter finden sich auch auch einige deren oriCs schon identifiziert worden sind

Der Folgende Code Block läd zwei Textdateien aus dem Internet herunter auf die wir dann programmatisch zugreifen können, dabei handelt es sich um den Forward und Reverse half-strand von Thermotoga petrophila.


```
#r "nuget: FSharp.Data, 6.0.1-beta002"
open FSharp.Data
let tpForwardRaw = Http.RequestString(@"https://raw.githubusercontent.com/jimandreas/stepikBioinformaticsCourse/main/src/main/resources/EcoliGenome.txt")
let tpReverseRaw = Http.RequestString(@"https://raw.githubusercontent.com/jimandreas/stepikBioinformaticsCourse/main/src/main/resources/EcoliGenome.txt")
```

Schauen wir uns an wie die ersten 100 Zeichen unseres forward strands aussehen:

```
Seq.take 100 forwardRaw
```

ja, das sieht tatsächlich nach einer DNA-Sequenz aus, schauen wir uns an wie es um die Gesamtlänge steht:

```
forwardRaw.Length
```

ungefähr neunhundertausend Basen. Auch das ist korrekt.

Nun ist es etwas unhandlich auf diese Weise mit einer Genom Sequenz zu hantieren. 
Um die Sache etwas angenehmer zu gestalten verwenden wir nun eine Code-Bibliothek mit dem Namen Biofsharp. Diese besitzt bereits einge Grundfunkionalitäten im Umgang mit Biologischen Sequenzen

```
#r "nuget: BioFSharp.Interactive, 2.0.0-preview.3"
open BioFSharp

let tpForward = BioSeq.ofNucleotideString tpForwardRaw
let tpReverse = BioSeq.ofNucleotideString tpReverseRaw
```

Zum Beispiel die schönere Formatierung unserer Ergebnisse:

```
Seq.take 500 thermotogaPetrophilaGenome
```

Nach unserem bisherigen Verständnis sollten die Anzahl der jeweiligen Basen Beispielsweise zwischen Reverse und Forward strands identisch sein. Das können wir nun überprüfen indem wir eine Statistik erstellen und visualisieren:

Zunächst Zählen wir die Anzahl einzelner Nukleotide im Forward Half strand

```
let anzahlForward_A =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.A)
    |> Seq.length

let anzahlForward_T =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.T)
    |> Seq.length

let anzahlForward_G =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.G)
    |> Seq.length

let anzahlForward_C =
    tpForward 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.C)
    |> Seq.length
```

```
let anzahlReverse_A =
    tpReverse 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.A)
    |> Seq.length

let anzahlReverse_T =
    tpReverse 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.T)
    |> Seq.length

let anzahlReverse_G =
    tpReverse 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.G)
    |> Seq.length

let anzahlReverse_C =
    tpReverse 
    |> Seq.filter (fun x -> x = BioFSharp.Nucleotides.C)
    |> Seq.length
```
